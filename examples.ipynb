```python
import matplotlib.pyplot as plt
import SEDtools
```

```python
sed, fit = 'example_SEDs/magphys/magphys.sed', 'example_SEDs/magphys/magphys.fit'
magphys_example_SED = SEDtools.SEDProperties(sed, fit, 'spec', 'magphys')

fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
magphys_example_SED.plot_SED(ax, 'magphys', plot_RF = False, obs_ms = 7, xmax = 1E3, ymin = -2, ymax = 3.5)
```

```python
fig = plt.figure(figsize=(20, 5))

ax = plt.subplot(121)
sed, fit = 'example_SEDs/sed3fit/sed3fit.sed', 'example_SEDs/sed3fit/sed3fit.fit'
s3f_example_SED = SEDtools.SEDProperties(sed, fit, 'spec', 'sed3fit')
s3f_example_SED.plot_SED(ax, 'S3F', obs_ms = 7, \
                         plot_dustySF = True, plot_StUn = True, plot_RF = False, \
                         xmax = 4E3, ymin = -7, ymax = -1)

sed, fit = 'example_SEDs/sed3fit/with_uplims.sed', 'example_SEDs/sed3fit/with_uplims.fit'
s3f_example_SED = SEDtools.SEDProperties(sed, fit, 'spec', 'sed3fit')
ax = plt.subplot(122)
s3f_example_SED.plot_SED(ax, 'S3F with uplims', \
             plot_RF = True, xmax = 1E4, ymin = -9.1, ymax = 0.5)
```

```python
sed, fit = 'example_SEDs/magphys/magphys_photoz.sed', 'example_SEDs/magphys/magphys_photoz.fit'
magphysphotoz_example_SED = SEDtools.SEDProperties(sed, fit, 'phot', 'magphys')

fig = plt.figure(figsize=(22,5))
ax = plt.subplot(121)
magphysphotoz_example_SED.plot_SED(ax, 'magphys+photoz', SED_observed_color = 'tab:cyan', plot_RF = True, obs_ms = 7, xmax = 1E3, ymin = -8, ymax = 0)

ax = plt.subplot(122)
magphysphotoz_example_SED.plot_SED(ax, 'magphys+photoz, shorter inset', SED_observed_color = 'magenta', \
                                   z_inset_xlength = 0.4, zphot_text_fontsize=10, zphot_inset_xmin = 0, zphot_inset_xmax = 2, zphot_text_xtext = 0.1, \
                                   plot_RF = True, obs_ms = 7, xmax = 1E3, ymin = -8, ymax = 0)
```
